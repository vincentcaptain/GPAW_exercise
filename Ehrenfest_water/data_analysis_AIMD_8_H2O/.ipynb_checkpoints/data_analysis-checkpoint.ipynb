```python
from pylab import *
```

```python
time_frame = around(arange(0.1, 0.6, 0.1), decimals = 2)
print(time_frame)
```

```python
OO_init = loadtxt("O-O_init.dat")[:, 1][11:]
OH_init = loadtxt("O-H_init.dat")[:, 1][11:]
HH_init = loadtxt("H-H_init.dat")[:, 1][11:]
R_range = loadtxt("O-O_init.dat")[:, 0][11:]
print(len(R_range))
init_signal = OO_init + OH_init + HH_init
plot(R_range, init_signal)
```

```python
OO_signal = []
OO_signal_diff = []
OH_signal = []
OH_signal_diff = []
HH_signal = []
HH_signal_diff = []
combined_signal = []
combined_signal_diff = []
for t in time_frame:
    OO = loadtxt("O-O_%s.dat" % t)[:, 1][11:]
    OH = loadtxt("O-H_%s.dat" % t)[:, 1][11:]
    HH = loadtxt("H-H_%s.dat" % t)[:, 1][11:]
    combined = OO + OH + HH
    OO_signal.append(OO)
    OO_signal_diff.append(OO - OO_init)
    OH_signal.append(OH)
    OH_signal_diff.append(OH - OH_init)
    HH_signal.append(HH)
    HH_signal_diff.append(HH - HH_init)
    combined_signal.append(combined)
    combined_signal_diff.append(combined - init_signal)

OO_signal = asarray(OO_signal)
OO_signal_diff = asarray(OO_signal_diff)
OH_signal = asarray(OH_signal)
OH_signal_diff = asarray(OH_signal_diff)
HH_signal = asarray(HH_signal)
HH_signal_diff = asarray(HH_signal_diff)
combined_signal = asarray(combined_signal)
combined_signal_diff = asarray(combined_signal_diff)
```

```python
OO_signal.shape
```

```python
p = contourf(R_range, time_frame, OO_signal_diff, 12)
xlabel("Radius", fontsize = 18)
ylabel("delayed time", fontsize=18)
rc("xtick", labelsize=15)
rc('ytick', labelsize=15)
title("O-O RDF contour")
```

```python
p = contourf(R_range, time_frame, OH_signal_diff, 10)
xlabel("Radius", fontsize = 18)
ylabel("delayed time", fontsize=18)
rc("xtick", labelsize=15)
rc('ytick', labelsize=15)
title("O-H RDF contour")
```

```python
contourf(R_range, time_frame, HH_signal_diff, 10)
xlabel("Radius", fontsize = 18)
ylabel("delayed time", fontsize=18)
rc("xtick", labelsize=15)
rc('ytick', labelsize=15)
title("H-H RDF contour")
```

```python
contourf(R_range, time_frame, combined_signal_diff, 10)
xlabel("Radius", fontsize = 18)
ylabel("delayed time", fontsize=18)
rc("xtick", labelsize=15)
rc('ytick', labelsize=15)
title("Combined signal contour")
```
